```python
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generateRandomTS(x):
    rTS=[]
    rts=0
    for i in range(0,len(x)):
        if i<len(x)/5:
            rts=np.sqrt(x[i]/10-1)
            rTS.append(rts+np.random.uniform(-0.2,0.2))
        else:
            rTS.append(np.random.uniform(-0.2,0.2)+rts)
    return np.array(rTS)
```

```python
x=np.linspace(10,1000,2000)
myrTS=generateRandomTS(x)
plt.plot(myrTS)
plt.show()
```

```python
def SSD1(inVec,BinNum,PeReM,PeReV):
    steady_state=1
    lVec=len(inVec)
    lBin=lVec//BinNum
    #if len(inVec)%2==0 and :
    meanVec=[]
    varVec=[]
    for i in range(BinNum):
        jBeg=i*lBin
        jEnd=(i+1)*lBin
        meanVec.append(np.mean(inVec[jBeg:jEnd]))
        varVec.append(np.var(inVec[jBeg:jEnd]))
        #myVec.append(np.array(inVec[jBeg:jEnd]))
    #all_mean=np.mean(np.array(meanVec))
    #all_var=np.var(np.array(varVec))
    m0=meanVec[0]
    v0=varVec[0]
    for i in range(BinNum):
        if abs(m0-meanVec[i])>PeReM or abs(v0-varVec[i])>PeReV:
            steady_state=0
    return steady_state
```

```python
l0,l1=100,200
print("steady state ? ",SSD1(myrTS[l0:l1],10,0.1,0.1))
plt.plot(myrTS[l0:l1])
plt.show()
```

```python
l0,l1=200,400
print("steady state ? ",SSD1(myrTS[l0:l1],10,0.1,0.1))
plt.plot(myrTS[l0:l1])
plt.show()
```
